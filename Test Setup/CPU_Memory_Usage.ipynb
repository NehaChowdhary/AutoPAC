```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('CPU_RAM_Metric.csv')
```

```python
df.head()
```

```python
max_cpu_usage = df['CPU_Usage'].max()
avg_cpu_usage = df['CPU_Usage'].mean()
min_cpu_usage = df['CPU_Usage'].min()

max_memory_usage = df['Memory_Usage'].max()
avg_memory_usage = df['Memory_Usage'].mean()
min_memory_usage = df['Memory_Usage'].min()

max_memory_percent = df['Memory_Percentage'].max()
avg_memory_percent = df['Memory_Percentage'].mean()
min_memory_percent = df['Memory_Percentage'].min()
```

```python
print("Max. CPU Usage: {}%".format(max_cpu_usage))
print("Min. CPU Usage: {}%".format(min_cpu_usage))
print("Avg. CPU Usage: {}%".format(avg_cpu_usage))

print("-"*50)

print("\nMax. Memory Usage: {} GB".format(max_memory_usage))
print("Min. Memory Usage: {} GB".format(min_memory_usage))
print("Avg. Memory Usage: {} GB".format(avg_memory_usage))

print("\nTotal Memory Available: {} GB".format(df['Total_Memory'].mean()))

print("-"*50)

print("\nMax. Memory Percentage: {} %".format(max_memory_percent))
print("Min. Memory Usage: {} %".format(min_memory_percent))
print("Avg. Memory Usage: {} %".format(avg_memory_percent))
```

```python
plt.plot(df['CPU_Usage'])
plt.xlabel('Number of Rules')
plt.ylabel('CPU Usage in %')
plt.title('CPU Usage during PrivateGPT querying')
```

```python
plt.plot(df['Memory_Usage'])
plt.xlabel('Number of Rules')
plt.ylabel('Memory Usage in GB')
plt.title('Memory Usage during PrivateGPT querying')
```

```python
plt.plot(df['Memory_Percentage'])
plt.xlabel('Number of Rules')
plt.ylabel('Memory Usage in %')
plt.title('Memory Usage during PrivateGPT querying')
```
